```python
# Importing the packages I will use in this project

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None



# Reading in the data

df = pd.read_csv(r'C:\Users\ronjo\Documents\Data Analytics\Python Projects\movies.csv')
```

```python
# Looking at the data

df.head()
```

```python

# Looking for missing data by Loop through


for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# Looking at Data Types for the columns

print(df.dtypes)
```

```python
# Are there any Outliers?

df.boxplot(column=['gross'])
```

```python
# Dropping duplicates

df.drop_duplicates()
```

```python
# Ordering the Data

df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# Perceiving that movies with a high budget will correlate with a high gross.
# Plotting the budget vs the gross using seaborn

sns.regplot(x="gross", y="budget", data=df, scatter_kws={"color": "red"}, line_kws={"color":"blue"})
```

```python
# Checking the correlation

df.corr()
```

```python
# Visualizing the correlation matrix

correlation_matrix = df.corr()

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Numeric Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
# Using factorize - this assigns a random numeric value for each unique categorical value

df.apply(lambda x: x.factorize()[0]).corr(method='pearson')
```

```python
correlation_matrix = df.apply(lambda x: x.factorize()[0]).corr(method='pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Movies")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
correlation_mat = df.apply(lambda x: x.factorize()[0]).corr()

corr_pairs = correlation_mat.unstack()

print(corr_pairs)
```

```python

sorted_pairs = corr_pairs.sort_values(kind="quicksort")

print(sorted_pairs)
```

```python
# Plotting high correlation (> 0.5)

strong_pairs = sorted_pairs[abs(sorted_pairs) > 0.5]

print(strong_pairs)
```

```python
# Looking at the top 15 compaies by gross revenue

CompanyGrossSum = df.groupby('company')[["gross"]].sum()

CompanyGrossSumSorted = CompanyGrossSum.sort_values('gross', ascending = False)[:15]

CompanyGrossSumSorted = CompanyGrossSumSorted['gross'].astype('int64') 

CompanyGrossSumSorted
```
